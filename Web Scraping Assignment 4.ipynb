1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views



```python
def most_viewed_videos(url):
    
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    ranks = []
    video_names = []
    artists = []
    dates = [] 
    views = []

    try:
        rank = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[1]")
        for r in rank:
            ranks.append(r.text)
    except NoSuchElementException :
        ranks.append(" - ")

    try:
        videos = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[2]")
        for vi in videos:
            video_names.append(vi.text.split("[")[0])
    except NoSuchElementException :
        video_names.append(" - ")

    try:
        artist = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[3]")
        for a in artist:
            artists.append(a.text)
    except NoSuchElementException :
        artists.append(" - ")

    try:
        view = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[4]")
        for v in view:
            views.append(v.text)
    except NoSuchElementException :
        views.append(" - ")

    try:
        date = driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[5]")
        for d in date:
            dates.append(d.text)
    except NoSuchElementException :
        dates.append(" - ")

    top_videos = pd.DataFrame({'Rank' : ranks,
                               'Video Name' : video_names,
                               'Uploader/artist' : artists,
                              'Views ( billions )' : views,
                              'Upload Date' : dates})
    driver.close()
    
    return top_videos

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
df = most_viewed_videos(url)
df
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.


```python
def international_fixtures(url):
    
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)


    #goToInternationalFixtures
    fixture = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a")
    fixture_link = fixture.get_attribute('href')
    driver.get(fixture_link)

    match_titles = []
    series = []
    places = []
    dates = []
    times = []

    try:
        game_urls = driver.find_elements_by_xpath("//a[contains(@href,'1st-odi')]")
        for i in game_urls:
            driver.get(i.get_attribute('href'))

            series_name = driver.find_element_by_xpath("//*[@id='scorecard']/section[3]/div/ul/li[1]/span[2]")
            series.append(series_name.text)

            title = driver.find_element_by_xpath("/html/body/div[2]/div/header/div/div[2]/div/p/span")
            match_titles.append(title.text)

            venue = driver.find_element_by_xpath("//*[@id='scorecard']/section[3]/div/ul/li[3]/span[2]")
            places.append(venue.text)

            date = driver.find_element_by_xpath("/html/body/div[2]/div/header/div/div[3]/div/div[3]/strong")
            dates.append(date.text)

            time = driver.find_element_by_xpath("/html/body/div[2]/div/header/div/div[3]/div/div[3]/span[1]")
            times.append(time.text)

    except NoSuchElementException:
        print("None")
    driver.close()
    odi_matches = pd.DataFrame({'Match Title': match_titles,
                                'Series Name': series,
                                'Date': dates,
                                'Time':times,
                                'Venue':places})

    return odi_matches

url = "https://www.bcci.tv"
df = international_fixtures(url)
df
```

```python
def selenium_exception(url):
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    selenium = driver.find_element_by_xpath("//a[@title='Selenium']")
    driver.get(selenium.get_attribute('href'))


    tutorial = driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']")
    driver.get(tutorial.get_attribute('href'))

    #WebDriverWait(driver,60).until(EC.element_to_be_clickable((By.XPATH,"//button[@class='CloseButton__ButtonElement-sc-79mh24-0 gVNUOF aurora-CloseButton aurora-close aurora-ClosePosition--top-right']"))).click()


    names = []
    descriptions = []


    try:
        name = driver.find_elements_by_xpath("//*[@id='g-mainbar']/div[1]/div/div/div/div/div/div[2]/table/tbody/tr/td[1]")
        for n in name:
            names.append(n.text)
    except NoSuchElementException :
        names.append(" - ")

    try:
        desc = driver.find_elements_by_xpath("//*[@id='g-mainbar']/div[1]/div/div/div/div/div/div[2]/table/tbody/tr/td[2]")
        for d in desc:
            descriptions.append(d.text)
    except NoSuchElementException :
        descriptions.append(" - ")  


    exceptions = pd.DataFrame({'Name':names,
                              'Description':descriptions
                              })
    driver.close()
    exceptions.drop(axis=0,index=0,inplace=True)
    
    return exceptions

url = " https://www.guru99.com"
df = selenium_exception(url)
df
```

4. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details:
A) Rank
B) State
C) GSDP(18-19)
D) GSDP(17-18)
E) Share(2017)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code

```python
def gdp(url):

    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)


    india = driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/div/a[3]")
    driver.get(india.get_attribute('href'))

    gdp = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/ul/li[2]/a")
    driver.get(gdp.get_attribute('href'))

    ranks = []
    states = []
    gsdp_18_19_cur = []
    gsdp_19_20_cur = []
    gsdp_19_20_con = []
    gsdp_18_19_con = []
    gdp_billion = []

    try:
        rank = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]")
        for r in rank:
            ranks.append(r.text)
    except NoSuchElementException:
        ranks.append("-")

    try:
        state = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]")
        for s in state:
            states.append(s.text)
    except NoSuchElementException:
        states.append("-")

    try:
        gsdp_18_cur = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
        for g in gsdp_18_cur:
            gsdp_18_19_cur.append(g.text)
    except NoSuchElementException:
        gsdp_18_19_cur.append("-")

    try:
        gsdp_19_cur = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]")
        for s in gsdp_19_cur:
            gsdp_19_20_cur.append(s.text)
    except NoSuchElementException:
        gsdp_19_20_cur.append("-")

    try:
        billion = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
        for b in billion:
            gdp_billion.append(b.text)
    except NoSuchElementException:
        ranks.append("-")

    try:
        gsdp_18_con = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[7]")
        for d in gsdp_18_con:
            gsdp_19_20_con.append(d.text)
    except NoSuchElementException:
        gsdp_19_20_con.append("-")

    try:
        gsdp_19_con = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[8]")
        for p in gsdp_19_con:
            gsdp_18_19_con.append(p.text)
    except NoSuchElementException:
        gsdp_18_19_con.append("-")
    
    driver.close()
    
    gdp_indian_states = pd.DataFrame({'Rank':ranks,
                                      'State':states,
                                      'GSDP Per Capita (INR at current Prices) - 2018-19':gsdp_18_19_cur,
                                      'GSDP Per Capita (INR at current Prices) - 2019-20':gsdp_18_19_cur,
                                      'GSDP Per Capita (INR at constant Prices) - 2018-19':gsdp_18_19_con,
                                      'GSDP Per Capita (INR at constant Prices) - 2019-20':gsdp_19_20_con,
                                      'GDP Per capita ($)': gdp_billion
                                     })

    return gdp_indian_states

url = "http://statisticstimes.com"
df = gdp(url)
df
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
def trending_repo(url):
    
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    trending = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
    driver.get(trending.get_attribute('href'))

    repo_urls = []
    repo_title = []
    desc = []
    contributors_count = []
    languages = []


    try:
        repositories = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
        for r in repositories:
            repo_urls.append(r.get_attribute('href'))
            repo_title.append(r.get_attribute('href').split("com/")[1])
    except NoSuchElementException:
        repo_title.append("--")
        repo_urls.append("--")

    for url in repo_urls:
        driver.get(url)
        try:
            about = driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div[2]/div[2]/div/div[1]/div/p")
            desc.append(about.text)
        except NoSuchElementException:
            desc.append(" -- ")
        try:
            counter = driver.find_element_by_xpath("//a[contains(@href, '/contributors')]/span")
            contributors_count.append(counter.get_attribute('innerHTML'))
        except NoSuchElementException:
            contributors_count.append(" -- ")

        try:
            lang = driver.find_elements_by_xpath("//*[@id='repo-content-pjax-container']/div[2]/div[2]/div/div/div/ul/li/a/span[1]")
            list_of_lang = []
            for l in lang:
                list_of_lang.append(l.text)
            languages.append(list_of_lang)

        except NoSuchElementException:
            languages.append(" -- ")
    
    driver.close()
    
    trending_posts = pd.DataFrame({'Repository Title': repo_title,
                                   'Repository Description': desc,
                                   'Contributors Count': contributors_count,
                                   'Languages Used': languages
                                  })
    return trending_posts

url = "https://github.com"
df = trending_repo(url)
```

6. Scrape the details of top 100 songs on billiboard.com.
Url = https://www.billiboard.com/
You have to find the following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board
Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
def top_songs(url):
    
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    charts = driver.find_element_by_xpath("//a[@class='header__main-link header__main-link--charts']")
    driver.get(charts.get_attribute('href'))


    hot_100 = driver.find_element_by_xpath("//div[@id='topchartsChartPanel']/a")
    driver.get(hot_100.get_attribute('href'))

    song_names = []
    artist_names = []
    last_week_ranks = []
    peak_ranks = []
    weeks_on_charts = []

    try:
        songs = driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
        for s in songs:
            song_names.append(s.text)
    except NoSuchElementException:
        song_names.append(" - ")

    try:
        artists = driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
        for a in artists:
            artist_names.append(a.text)
    except NoSuchElementException:
        artist_names.append(" - ")

    try:
        week = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
        for w in week:
            last_week_ranks.append(w.text)
    except NoSuchElementException:
        last_week_ranks.append(" - ")

    try:
        peak = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
        for p in peak:
            peak_ranks.append(p.text)
    except NoSuchElementException:
        peak_ranks.append(" - ")

    try:
        charts = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
        for c in charts:
            weeks_on_charts.append(c.text)
    except NoSuchElementException:
        weeks_on_charts.append(" - ")

    song_list = pd.DataFrame({'Song Name': song_names,
                              'Artist Name': artist_names,
                              'Last Week Rank' : last_week_ranks,
                              'Peak Rank': peak_ranks,
                              'Weeks on Board' : weeks_on_charts})
    driver.close()
    return song_list

url = "https://www.billboard.com/"
df = top_songs(url)
df
```

D) Skills they hire for

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and
click on search. All this should be done through code

```python
#Function definition
def find_jobs_from(url) :
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    recruiters = driver.find_element_by_xpath("//a[@title='Search Recruiters']")
    driver.get(recruiters.get_attribute('href'))
                                            
    skill_field = driver.find_element_by_name('qp')
    skill_field.send_keys("Data Scientist")


    search_button = driver.find_element_by_xpath("//button[@class='fl qsbSrch blueBtn']")
    try:
        search_button.click()
    except ElementNotInteractableException:
        print("Search button Not responding!!")

    current_url = driver.current_url

    driver.get(current_url)
    print(current_url)
    names = []
    locations = []
    company_name = []
    designation = []
    skills = []

    try:                                                        
        name = driver.find_elements_by_xpath("//div[@id='tabP-1']/div/div/div[1]/div[1]/p/a[1]/span")
        for i in name:
            names.append(i.get_attribute('innerHTML'))
    except NoSuchElementException:
        names.append(" -- ")

    
                                               
    location = driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/span[2]/small")
    for j in location:
        if j.text is None:
             locations.append(" -- ")
        else:
            locations.append(j.text)

    try:                                         
        company = driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/a[2]")
        for l in company:
            company_name.append(l.text)
    except NoSuchElementException:
        company_name.append(" -- ")
                
    try:                                         
        designations = driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/span[1]")
        for m in designations:
            designation.append(m.text)
    except NoSuchElementException:
        designation.append(" -- ")
        
    try:                                         
        skill = driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
        for n in skill:
            skills.append(n.text)
    except NoSuchElementException:
        skills.append(" -- ")
   
    driver.close()
    data_analyst_jobs = pd.DataFrame({"Name" : names[0:49],
                                      'Designation': designation[0:49],
                                      'Location' : locations[0:49],
                                      'Company' : company_name[0:49],
                                      'Skills they Hire' : skills[0:49]
                                     })
    return data_analyst_jobs

#Call Function
df = pd.DataFrame()
df = find_jobs_from("https://www.naukri.com/")
df
```

You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
def most_watched_series(url):
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    name = []
    year_span = []
    genre = []
    run_time = []
    ratings = []
    votes = []

    try:
        names = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/h3/a")
        for n in names:
            name.append(n.text)
    except NoSuchElementException:
        name.append(" -- ")

    try:
        year = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/h3/span[2]")
        for y in year:
            year_span.append(y.text)
    except NoSuchElementException:
        year_span.append(" -- ")

    try:
        genres = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/p[1]/span[5]")
        for g in genres:
            genre.append(g.text)
    except NoSuchElementException:
        genre.append(" -- ")

    try:
        runtime = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/p[1]/span[3]")
        for r in runtime:
            run_time.append(r.text)
    except NoSuchElementException:
        run_time.append(" -- ")

    try:
        rating = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/div[1]/div[1]/span[2]")
        for ra in rating:
            ratings.append(ra.text)
    except NoSuchElementException:
        ratings.append(" -- ")

    try:
        vote = driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/p[4]/span[2]")
        for v in vote:
            votes.append(v.text)
    except NoSuchElementException:
        votes.append(" -- ")

    top_Serial_List = pd.DataFrame({'Name': name,
                                    'Year Span' : year_span,
                                    'Genre ': genre,
                                    'Run Time' : run_time,
                                    'Ratings': ratings,
                                    'Votes': votes})
    driver.close()
    return top_Serial_List

url = "https://www.imdb.com/list/ls095964455"
df = most_watched_series(url)
df
```

10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the ShowAllDataset page through code.

```python
def uci_ml_repo(url):
    driver = webdriver.Edge("msedgedriver.exe")
    driver.get(url)

    all_dataset = driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a")
    driver.get(all_dataset.get_attribute('href'))

    set_name = []
    data_type = []
    task = []
    attribute_type = []
    no_of_instance = []
    no_of_attribute = []
    year = []


    try:                                         
        dataset = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td/table/tbody/tr/td/p/b/a")
        for d in dataset:
            set_name.append(d.text)
    except NoSuchElementException:
        set_name.append(" -- ")


    try:                                                                                   
        datatype = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
        for d in datatype[1:]:
            data_type.append(d.text)
    except NoSuchElementException:
        set_name.append(" -- ")

    try:                                                                                   
        tasks = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
        for t in tasks[1:]:
            task.append(t.text)
    except NoSuchElementException:
        task.append(" -- ")

    try:                                                                                   
        attributetype = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
        for a in attributetype[1:]:
            attribute_type.append(a.text)
    except NoSuchElementException:
        attribute_type.append(" -- ")

    try:                                                                                   
        instances = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
        for i in instances[1:]:
            no_of_instance.append(i.text)
    except NoSuchElementException:
        no_of_instance.append(" -- ")

    try:                                                                                  
        attributes = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
        for at in attributes[1:]:
            no_of_attribute.append(at.text)
    except NoSuchElementException:
        no_of_attribute.append(" -- ")

    try:                                                                                 
        years = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
        for y in years[1:]:
            year.append(y.text)
    except NoSuchElementException:
        year.append(" -- ")

    driver.close()
    uci_datasets = pd.DataFrame({'Dataset Name': set_name,
                                 'Data Type': data_type,
                                 'Task': task,
                                 'Attribute Type': attribute_type,
                                 'Number of Instances': no_of_instance,
                                 'Number of Attribute': no_of_attribute,
                                 'Year': year
                                })

    return uci_datasets

url = "https://archive.ics.uci.edu"
df = uci_ml_repo(url)
df
```
