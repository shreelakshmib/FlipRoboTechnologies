```python
from bs4 import BeautifulSoup as BS
import requests
```

```python
page = requests.get("https://in.indeed.com/data-scientist-jobs-in-Bengaluru,-Karnataka")
page.content
```

```python
soup = BS(page.content,'html.parser')
```

```python
job_titles = []
company_names = []
locations_list = []
days_posted = []
```

```python
titles = soup.find_all('a',class_ = 'jobtitle turnstileLink')
titles[0:4]
```

```python
for i in titles:
    job_titles.append(i.get_text().replace("\n",""))
job_titles
```

```python
companies = soup.find_all('span',class_ = 'company')
companies[0:4]
```

```python
for i in companies:
    company_names.append(i.get_text().replace("\n",""))
company_names
```

```python
locations = soup.find_all('span',class_ = 'location accessible-contrast-color-location')
locations[0:4]
```

```python
for i in locations:
    locations_list.append(i.get_text())
locations_list
```

```python
days = soup.find_all('span',class_ = 'date')
days[0:4]
```

```python
for i in days:
    days_posted.append(i.get_text())
days_posted
```

```python
print(len(job_titles),len(company_names),len(locations_list),len(days_posted))
```

```python
import pandas as pd
jobs = pd.DataFrame({})
jobs['title'] = job_titles
jobs['company name'] = company_names
jobs['location'] =locations_list
jobs['posted day'] = days_posted

jobs
```

```python
jobs.to_csv('jobs.csv')
```
